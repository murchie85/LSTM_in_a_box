````python
import spacy

# Load language model. 
nlp = spacy.load('en', disable = ['parser', 'tagger', 'ner'])
````

# Extract Tokens (words)
if the data set is big, it might be necessary to increase nlp.max_length.  
Remember to run python -m spacy download en to allow en function if not working 

````python
def get_tokens(doc_text):
    # This pattern is a modification of the defaul filter from the
    # Tokenizer() object in keras.preprocessing.text. 
    # It just indicates which patters no skip.
    skip_pattern = '\r\n \n\n \n\n\n!"-#$%&()--.*+,-./:;<=>?@[\\]^_`{|}~\t\n\r '
    
    tokens = [token.text.lower() for token in nlp(doc_text) if token.text not in skip_pattern]
    
    return tokens
````

````python
# Get tokens.
tokens = get_tokens(fullData)
````

````python
# Let us see the first tokens.
tokens[0:9]
````

````python
# Compute the number of tokens list.
print('Length of processed tokens are : ' + str(len(tokens) ))
````

# SEQUENCE GENERATION FOR FEATURE EXTRACTION 
 
We build a sequence of words in order to predict the next word.  
len_0 = feature (what we know)  
len_0 + 1 = target (or eventually the label what we want to predict)    

The model used will be a multi-class classification problem. 

````python
len_0 = 25
print('Tokens/feature example is: \n'+ str(tokens[0:len_0]))
print('')
print('Target or label is ' + str(tokens[len_0:len_0 + 1]))
````

````python
### Generate sequence 
````

````python
train_len = len_0 + 1

text_sequences = []

for i in range(train_len, len(tokens)):
    # Construct sequence.
    seq = tokens[i - train_len: i]
    # Append.
    text_sequences.append(seq)
````

````python
print(' '.join(text_sequences[0]))
print('Len of sequence is : ' + str(len(text_sequences[0])))
````

````python
# First three sequences:
for i in range(0, 3):
    print(' '.join(text_sequences[i]))
    print('****')
````

# Words to vectors (Vectorisation) 

- fit on ur sequence array of 26 words per element
- Now the word seuqnces will be converted `(encoded)` as `numerical features` : the keras `Tokenizer` object does this.

````python
from keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()

tokenizer.fit_on_texts(text_sequences)
````

````python
# Get numeric sequences.
sequences = tokenizer.texts_to_sequences(text_sequences)


print('len of sequence is still ' + str(len(sequences[0])))
print(sequences[0])
print('First char index ' + str(sequences[0][0]) + ' should map to the first char in the token string' )
````

````python
# Verify  the above is correct
print(tokenizer.index_word[int(sequences[0][0])])
print(fullData.split(' ')[0])
````

````python
# Vocab size is the number of unique tokens
vocabulary_size = len(tokenizer.word_counts)
print('vocab size ' + str(vocabulary_size))
````

````python
# Pass sequences to np array so we can use it in our xy split
sequences = np.array(sequences)
print(sequences)
````

# X - y Split
We now construct the observation matrix X and the label vector y.   
(X is all 25  words of each elm, y is the 26th)  
Observation Matrix Definition: `It is a measure of how dependent the measurements are upon the state of the system`

````python
from keras.utils import to_categorical
# select all but last word indices.
X = sequences[:, :-1]
X
````

````python
X.shape
````

````python
seq_len = X.shape[1] # assigning sequence len to second parm of shape array
````

````python
# select all last word indices.
y = sequences[:, -1]
y
````

````python
# Covert to categorical (we add + 1 because Keras needs a placeholder).
# Definition A categorical variable is a variable whose values take on the value of labels.
# Encoding is possibly one hot encoding binary values of 0 or 1
y = to_categorical(y, num_classes=(vocabulary_size + 1))
y
````

# MODEL ARCHITECTURE
  
- Sequential network
- Embedding layer
- Two LSTM layers 
- One Dense layer with RELU activation function
- Final softmax Dense function to output class probabilities 
  
### LSTM general design 
  
  
### For the optimization:

- loss = `categorical_crossentropy`
- optimizer = `adam’`
- metrics = ` [‘accuracy’] `

### Creating Model
`Sequential` `embedding` `LSTM` `LSTM` `Dense RELU` `Dense Softmax` 

````python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding

def create_model(vocabulary_size, seq_len):
    
    model = Sequential()

    model.add(Embedding(input_dim=vocabulary_size, 
                        output_dim=seq_len, 
                        input_length=seq_len))

    model.add(LSTM(units=50, return_sequences=True))

    model.add(LSTM(units=50))

    model.add(Dense(units=50, activation='relu'))

    model.add(Dense(units=vocabulary_size, activation='softmax'))

    model.compile(loss='categorical_crossentropy', 
                  optimizer='adam', 
                  metrics=['accuracy'])

    model.summary()
    return model
````

````python
# Let us create the model and see summary.
# Rememer we defined vocab size earlier (size of independent tokens) + 1 for placeholder
model = create_model(vocabulary_size=(vocabulary_size + 1), seq_len=seq_len)
````

# Generate New Text  

The below function strings together the most likely sequential words 

````python
from keras.preprocessing.sequence import pad_sequences 

def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    # List to store the generated words. 
    output_text = []
    # Set seed_text as input_text. 
    input_text = seed_text
    
    for i in range(num_gen_words):
        # Encode input text. 
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        # Add if the input tesxt does not have length len_0.
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        # Do the prediction. Here we automatically choose the word with highest probability. 
        pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0]
        # Convert from numeric to word. 
        pred_word = tokenizer.index_word[pred_word_ind]
        # Attach predicted word. 
        input_text += ' ' + pred_word
        # Append new word to the list. 
        output_text.append(pred_word)
        
    return ' '.join(output_text)
````

## Temperature  

Temperature is a hyperparameter of LSTMs (and neural networks generally) used to control the randomness of predictions by scaling the logits before applying softmax. For example, in TensorFlow’s Magenta implementation of LSTMs, temperature represents how much to divide the logits by before computing the softmax.  

Quote from cs exchange..  
```
When the temperature is 1, we compute the softmax directly on the logits (the unscaled output of earlier layers), and using a temperature of 0.6 the model computes the softmax on 𝑙𝑜𝑔𝑖𝑡𝑠0.6, resulting in a larger value. Performing softmax on larger values makes the LSTM more confident (less input is needed to activate the output layer) but also more conservative in its samples (it is less likely to sample from unlikely candidates). Using a higher temperature produces a softer probability distribution over the classes, and makes the RNN more “easily excited” by samples, resulting in more diversity and also more mistakes.
```

learning links  https://nlp.stanford.edu/blog/maximum-likelihood-decoding-with-rnns-the-good-the-bad-and-the-ugly/  

  

````python
def generate_text2(model, tokenizer, seq_len, seed_text, num_gen_words, temperature):
    
    output_text = []
    
    input_text = seed_text
    
    for i in range(num_gen_words):
        # Encode input text. 
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
         # Add if the input tesxt does not have length len_0.
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        # Get learned distribution.
        pred_distribution = model.predict(pad_encoded, verbose=0)[0]
        
        # Apply temperature transformation.
        new_pred_distribution = np.power(pred_distribution, (1 / temperature)) 
        new_pred_distribution = new_pred_distribution / new_pred_distribution.sum()
        
        # Sample from modified distribution.
        choices = range(new_pred_distribution.size)
 
        pred_word_ind = np.random.choice(a=choices, p=new_pred_distribution)
        
        # Convert from numeric to word. 
        pred_word = tokenizer.index_word[pred_word_ind]
        # Attach predicted word. 
        input_text += ' ' + pred_word
        # Append new word to the list. 
        output_text.append(pred_word)
    return ' '.join(output_text)
        
````

## TRYING WITH VARIOUS TEMP SETTINGS

````python
seed_text = 'I eat a lot of ass, the whole ass I love the whole ass'
temp = [0.9, 0.5, 0.1]
for tempValue in temp:
    print("Trying temperature at value : " + str(tempValue))
    generated_text = generate_text2(model=model, 
                                tokenizer=tokenizer,
                                seq_len=seq_len, 
                                seed_text=seed_text, 
                                num_gen_words=80, 
                                temperature=tempValue)

    print(str(seed_text) + ' ' + str(generated_text) + ' ...')
    print('')
    print('--------')
````
